```python
# Importación de las librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Semilla para asegurar reproducibilidad
np.random.seed(42)  # Establece una semilla para generar los mismos datos aleatorios cada vez que se ejecute el código

# Generar 100 estaturas entre 1.50 y 2.00 metros
estaturas = np.random.uniform(1.50, 2.00, 100)  # Genera 100 estaturas aleatorias en el rango especificado
estaturas = np.round(estaturas, 2)  # Redondea las estaturas a 2 decimales
```

```python
# Generar pesos basados en la estatura, asegurando coherencia
pesos = []
for estatura in estaturas:
    # Peso base es proporcional a la estatura más un poco de variabilidad usando la fórmula del IMC
    peso = np.random.uniform(18.5, 25.0) * (estatura ** 2)  # Calcula el peso en función de la estatura
    pesos.append(round(peso, 2))  # Redondea el peso a 2 decimales y lo agrega a la lista de pesos
```

```python
# Crear DataFrame con los datos generados
df = pd.DataFrame({
    'Estatura (m)': estaturas,
    'Peso (kg)': pesos
})


```

```python
# Calcular la pendiente (m) y la intersección (b) de la recta y = mx + b
x = df['Estatura (m)']  # Variable independiente (estaturas)
y = df['Peso (kg)']    # Variable dependiente (pesos)

# Calcula la pendiente (m) usando la fórmula de regresión lineal
m = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x)) ** 2)

# Calcula la intersección (b) de la línea de mejor ajuste
b = np.mean(y) - m * np.mean(x)

# Crear los valores de y basados en la fórmula de la recta
y_line = m * x + b
```

```python
# Visualización de los datos y la línea de ajuste
plt.scatter(df['Estatura (m)'], df['Peso (kg)'], color='blue', label='Datos')  # Muestra los puntos de datos
plt.plot(x, y_line, color='red', label='Línea ajustada')  # Muestra la línea de ajuste
plt.title('Estatura vs Peso con Línea Ajustada')  # Título del gráfico
plt.xlabel('Estatura (m)')  # Etiqueta del eje x
plt.ylabel('Peso (kg)')  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda del gráfico
plt.show()  # Muestra el gráfico
```
